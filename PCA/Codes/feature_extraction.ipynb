```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
load_breast_cancer()
```

```python
df = pd.DataFrame(data=load_breast_cancer().data, columns=load_breast_cancer().feature_names)
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

```python
pd.DataFrame(scaled_data)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca_data = pca.fit_transform(scaled_data)
```

```python
pca_data
```

```python
pca.explained_variance_
```

```python
pca = PCA()

pca.fit_transform(scaled_data)
```

```python
pca.explained_variance_ ## if we are not mentioning how many feature we want it is giving us variance for all the features
```

```python
## above two cells for just for eduction and PCA all the pca work done below is done with PCA(n_components=2)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(pca_data[:, 0], pca_data[:, 1], c=load_breast_cancer().target)
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
```
