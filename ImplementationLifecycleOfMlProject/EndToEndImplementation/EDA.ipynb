```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('../Datasets/Algerian_forest_fires_dataset.csv', header=1)
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df[df.isnull().any(axis=1)]
```

```python
df.loc[:122, 'Region'] = 0 # for bejiah region
df.loc[123:, 'Region'] = 1 # for sidi bel region
```

```python
df = df.dropna().reset_index(drop=True)
df.head()
```

```python
df.isna().sum()
```

```python
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop=True)
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip()
```

```python
df.columns
```

```python
int_type = ['day', 'month', 'year', 'Region']

for col in int_type:
    df[col] = df[col].astype('int')
```

```python
for col in df.columns:
    if col not in int_type and col != 'Classes':
        df[col] = df[col].astype('float')
```

```python
df.info()
```

```python
df.head()
```

```python
df['Classes'].value_counts()
```

```python
df['Classes'] = df['Classes'].str.strip()
```

```python
df['Classes'].value_counts()
```

```python
df['Classes'] = df['Classes'].map({
    "not fire": 0,
    "fire": 1
})
```

```python
df.head()
```

```python
df.to_csv('../Datasets/cleaned_alrerian_dataset.csv', index=False)
```
