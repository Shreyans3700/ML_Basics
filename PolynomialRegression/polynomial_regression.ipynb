```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Create dataset
X = 6 * np.random.rand(100, 1) - 3
y = 0.5 * X**2 + 1.5 * X + 2 + np.random.randn(100, 1)

# Quadratic eqn used - y = 0.5x^2 + 1.5x +  2 + outliers
```

```python
plt.scatter(X, y)
plt.xlabel('X')
plt.ylabel('y')
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression

regression = LinearRegression()
```

```python
regression.fit(X_train, y_train)
```

```python
y_pred = regression.predict(X_test)
```

```python
from sklearn.metrics import r2_score

scr = r2_score(y_pred, y_test)
print(scr)
```

```python
plt.plot(X_train, regression.predict(X_train), color='r')
plt.scatter(X_train, y_train, color='b')
plt.xlabel('X')
plt.ylabel('y')
```

```python
## Now let's try with Polynomial Regression
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=2, include_bias=True)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)
```

```python
regression.fit(X_train_poly, y_train)
```

```python
from sklearn.metrics import r2_score

scr = r2_score(regression.predict(X_test_poly), y_test)
print(scr)
```

```python
plt.scatter(X_train, regression.predict(X_train_poly), color='r')
plt.scatter(X_train, y_train, color='b')
plt.xlabel('X')
plt.ylabel('y')
```

```python
poly = PolynomialFeatures(degree=3, include_bias=True)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)

regression.fit(X_train_poly, y_train)

scr = r2_score(regression.predict(X_test_poly), y_test)
print(scr)
```
